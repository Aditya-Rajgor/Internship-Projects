```python
X=sample.loc[:,'MONTH':'WEATHER_DELAY']
```

```python
X.isna().sum()
```

```python
def inpute_mean(column):
    sample[column] = sample[column].fillna(sample[column].mean())
```

```python
inpute_mean('AIR_SYSTEM_DELAY')
inpute_mean('SECURITY_DELAY')
inpute_mean('AIRLINE_DELAY')
inpute_mean('LATE_AIRCRAFT_DELAY')
inpute_mean('WEATHER_DELAY')
inpute_mean('DEPARTURE_DELAY')
```

```python
y = sample.loc[:,'result']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train_scaled,y_train)
```
